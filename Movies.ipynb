```python
##drop columns 
Movie_drop = Movies_drop.drop(['released','name','star','writer','country','director'], axis = 1)
Movie_drop
```

```python
##First conclusions 
#1. Budget has a high positive correlation with gross
#2. High grossing movies generate more buzz, hence more votes
#3. More votes does not translate to a better score  
```

```python
##next step will be to measure to the degree each variable affects our dependent variable by running a regression
##drop some variables using Vif
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
variables = Movie_drop[['votes','budget','runtime','score']]
vif = pd.DataFrame()
vif["VIF"] = [variance_inflation_factor(variables.values, i) for i in range(variables.shape[1])]
vif["features"] = variables.columns
vif
```

```python
movie_clean = Movie_drop.drop(['runtime','log_gross'],axis=1)
movie_clean
```

```python
movies_dum = pd.get_dummies(movie_clean,columns=['genre'],drop_first = True)
movies_dum
```

```python
## for the regression analysis, I would like to assess the numerical impacts of 3 movie genres,(Drama,Fantasy,adventure,)
```

```python
movies_clean1= movies_dum.drop(['genre_Biography','genre_Western','company','genre_Family','genre_Horror'], axis =1)
movies_clean1
```

```python
movies_clean1.dtypes
```

```python
y= movies_clean1['gross']
X= movies_clean1.drop(['gross','rating'],axis =1)
x = sm.add_constant(X)
reg = sm.OLS(y,x).fit()
reg.summary()
```
